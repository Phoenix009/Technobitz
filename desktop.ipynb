```python
#Violinplot for tips using Matplotlib and Seaborn 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#SEABORN
#loading dataset
tips=sns.load_dataset("tips") 
#creating violinplot
ax=sns.violinplot(x="total_bill",data=tips)
#Show the plot
plt.show()
```

```python
#MATPLOTLIB
#Initializing figure and axes object
fig, ax = plt.subplots()
#loading dataset
tips = pd.read_csv("tips.csv")
#Plotting the violin plot
ax.violinplot(tips["total_bill"], vert=False)

plt.show()
```

```python
tips = sns.load_dataset("tips")
#Drawing a vertical violin plot by taking day variable on the X axis
ax = sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
#Grouping the data using different colours using hue- smoker vs non smoker 

ax=sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
# Drawing split violins to compare across the hue variables
ax=sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, split=True)
```

```python
#Drawing the quartiles using vertical lines instead of single box
ax=sns.violinplot(x="day",y="total_bill", hue="smoker", split=True, data=tips, inner="quartile")
```

```python
#Catplot using seaborn
# Used when one of the main variables is “categorical” (divided into discrete groups)
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
ax=sns.catplot(x="time", y="total_bill", hue="smoker", data=tips)
plt.show()
```

```python
#Divide catplot into columns to show third categorical variable
ax=sns.catplot(x="time", y="total_bill", hue="smoker",col="sex", data=tips)
plt.show()
```

```python
#Use a different kind of plot to show the same data
ax=sns.catplot(x="time", y="total_bill", hue="smoker", kind="violin", data=tips)
```

```python
#Countplot using Seaboen
#Shows the count of observations in each category using bars
ax=sns.countplot(x="time", hue="smoker", data=tips)
plt.show()
```

```python
#Plot the bars horizontally by taking the category on y axis
ax=sns.countplot(y="time", hue="smoker", data=tips)
```

```python
#Swarmplot using Seaborn
#It is a categorical scatterplot with non overlapping points
ax = sns.swarmplot(x=tips["total_bill"])
```
